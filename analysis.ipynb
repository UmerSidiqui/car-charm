```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
# Replace 'your_dataset.csv' with the path to your actual dataset file.
df = pd.read_csv('data.csv')

# Define speed bins (for example, 0-10, 10-20, etc.)
bins = range(0, 121, 10)  # Adjust the range and step as needed
labels = [f'{i}-{i+10}' for i in bins[:-1]]

# Create a new column for the speed bins
df['speed_bin'] = pd.cut(df['vehicle_speed'], bins=bins, labels=labels, include_lowest=True)

# Group by speed bins and calculate the average engine load for each bin
average_engine_load = df.groupby('speed_bin')['engine_rpm'].mean()

# Plot the bar graph
plt.figure(figsize=(10, 6))
average_engine_load.plot(kind='bar', color='blue', alpha=0.7)
plt.title('Average engine_rpm by Speed')
plt.xlabel('Speed (km/h)')
plt.ylabel('engine_rpm')
plt.xticks(rotation=45)
plt.show()
```
