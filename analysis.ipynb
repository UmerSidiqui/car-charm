```python
import pandas as pd
import plotly.graph_objects as go

# Assuming your dataset is loaded in the variable df

# Define the threshold for considering it a new trip
zero_speed_threshold = 5

# Create a column to identify long periods of zero speed
df['is_new_trip'] = (df['vehicle_speed'] == 0).astype(int)
df['consecutive_zeros'] = df['is_new_trip'].groupby((df['is_new_trip'] != df['is_new_trip'].shift()).cumsum()).cumsum()

# A new trip starts if consecutive zeros exceed the threshold
df['trip_id'] = ((df['consecutive_zeros'] > zero_speed_threshold) & (df['vehicle_speed'] == 0)).cumsum()

# Increment trip_id by 1 to ensure no 0 values
df['trip_id'] += 1

# Filter out rows where the vehicle speed is zero (they don't contribute to a trip)
trips_df = df[df['vehicle_speed'] > 0]

# Calculate top speed and average speed for each trip
trip_stats = trips_df.groupby('trip_id')['vehicle_speed'].agg(['max', 'mean']).reset_index()

# Rename the columns for clarity
trip_stats.columns = ['Trip ID', 'Top Speed (km/h)', 'Average Speed (km/h)']

# Visualize the results with horizontal lines
fig = go.Figure()

for i, row in trip_stats.iterrows():
    fig.add_trace(go.Scatter(
        x=[row['Trip ID'], row['Trip ID']],
        y=[row['Average Speed (km/h)'], row['Top Speed (km/h)']],
        mode='lines+markers',
        name=f'Trip {int(row["Trip ID"])}',
        line=dict(color='royalblue', width=2)
    ))

# Customize the layout
fig.update_layout(
    title='Top Speed and Average Speed for Each Trip',
    xaxis_title='Trip ID',
    yaxis_title='Speed (km/h)',
    showlegend=False,
    xaxis=dict(
        tickmode='linear',
        dtick=1
    )
)

# Add horizontal lines for average and top speeds
for i, row in trip_stats.iterrows():
    fig.add_shape(
        type='line',
        x0=row['Trip ID'] - 0.4,
        x1=row['Trip ID'] + 0.4,
        y0=row['Average Speed (km/h)'],
        y1=row['Average Speed (km/h)'],
        line=dict(color='green', width=2)
    )
    fig.add_shape(
        type='line',
        x0=row['Trip ID'] - 0.4,
        x1=row['Trip ID'] + 0.4,
        y0=row['Top Speed (km/h)'],
        y1=row['Top Speed (km/h)'],
        line=dict(color='red', width=2)
    )

fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px

# Assuming df is your dataframe with 'time_stamp' and 'vehicle_speed' columns
df['time_stamp'] = pd.to_datetime(df['time_stamp'])

# Create the base figure
fig = px.line(df, x='time_stamp', y='vehicle_speed', title='Vehicle Speed Over Time')
fig.update_traces(line=dict(color='blue'))

# Function to update annotations dynamically
def update_annotations(trace, points, selector):
    if len(points.point_inds) > 0:
        selected_points = points.point_inds
        selected_df = df.iloc[selected_points]
        
        # Calculate top speed and average speed for the selected range
        top_speed = selected_df['vehicle_speed'].max()
        avg_speed = selected_df['vehicle_speed'].mean()
        
        # Clear existing annotations
        fig.update_layout(annotations=[])
        
        # Add annotations for top speed and average speed
        fig.add_annotation(x=selected_df['time_stamp'].iloc[0], y=top_speed,
                           text=f"Top Speed: {top_speed:.2f} km/h", showarrow=True, arrowhead=1)
        fig.add_annotation(x=selected_df['time_stamp'].iloc[0], y=avg_speed,
                           text=f"Avg Speed: {avg_speed:.2f} km/h", showarrow=True, arrowhead=1)

# Connect the update function to the hover event
fig.data[0].on_selection(update_annotations)

# Show the figure
fig.show()
```

```python
import pandas as pd
import plotly.express as px
from dash import dcc, html, Dash, Input, Output

# Sample DataFrame creation


# Convert 'time_stamp' to datetime format
df['time_stamp'] = pd.to_datetime(df['time_stamp'])

# Create the Plotly figure
fig = px.line(df, x='time_stamp', y='vehicle_speed', title='Vehicle Speed Over Time')
fig.update_layout(
    xaxis_title='Time',
    yaxis_title='Vehicle Speed (km/h)',
    xaxis_rangeslider_visible=True
)

# Initialize the Dash app
app = Dash(__name__)

# Layout of the app
app.layout = html.Div([
    dcc.Graph(id='graph', figure=fig),
    html.Div(id='stats', style={'marginTop': 20})
])

# Callback to update the stats based on zoom/pan
@app.callback(
    Output('stats', 'children'),
    Input('graph', 'relayoutData')
)
def update_stats(relayoutData):
    if relayoutData and 'xaxis.range' in relayoutData:
        xaxis_range = relayoutData['xaxis.range']
        xaxis_range = [pd.to_datetime(x).to_pydatetime() for x in xaxis_range]
        zoomed_data = df[(df['time_stamp'] >= xaxis_range[0]) & (df['time_stamp'] <= xaxis_range[1])]
        avg_speed = zoomed_data['vehicle_speed'].mean()
        top_speed = zoomed_data['vehicle_speed'].max()
        return f'Average Speed: {avg_speed:.2f} km/h, Top Speed: {top_speed:.2f} km/h'
    return 'Adjust the zoom to see stats.'

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)


# Convert 'time_stamp' to datetime format
df['time_stamp'] = pd.to_datetime(df['time_stamp'])

# Create the Plotly figure
fig = px.line(df, x='time_stamp', y='vehicle_speed', title='Vehicle Speed Over Time')
fig.update_layout(
    xaxis_title='Time',
    yaxis_title='Vehicle Speed (km/h)',
    xaxis_rangeslider_visible=True
)

# Initialize the Dash app
app = Dash(__name__)

# Layout of the app
app.layout = html.Div([
    dcc.Graph(id='graph', figure=fig),
    html.Div(id='stats', style={'marginTop': 20})
])

# Callback to update the stats based on zoom/pan
@app.callback(
    Output('stats', 'children'),
    Input('graph', 'relayoutData')
)
def update_stats(relayoutData):
    if relayoutData and 'xaxis.range' in relayoutData:
        xaxis_range = relayoutData['xaxis.range']
        # Ensure that the timestamps are properly compared
        xaxis_range = [pd.to_datetime(x).to_pydatetime() for x in xaxis_range]
        zoomed_data = df[(df['time_stamp'] >= xaxis_range[0]) & (df['time_stamp'] <= xaxis_range[1])]
        avg_speed = zoomed_data['vehicle_speed'].mean()
        top_speed = zoomed_data['vehicle_speed'].max()
        return f'Average Speed: {avg_speed:.2f} km/h, Top Speed: {top_speed:.2f} km/h'
    return 'Adjust the zoom to see stats.'

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)

```
